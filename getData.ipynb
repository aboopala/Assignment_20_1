### Data Acquisition
This file pulls data from yfinance for the indicated tickers from Jan 2015 - Apr 2026. Modify start_date and end_date variables to pull a different subset of data.

Data is stored in the data folder as "stock_data.csv". Rerun the capstone to process a different subset of data. Note, the markdown cells will be associated with the dataset from Jan 2015 - Apr 2026.

```python
#Raw data previously downloaded and saved to CSV, so we can skip the download step in future runs# Choose a representative subset of large-cap U.S. stocks
import yfinance as yf 
import os
import pandas as pd


start_date = "2015-01-01"
end_date = "2026-04-22"  # For consistent data range across all tickers, use the same end date as the last download date

# Representative subset of large-cap U.S. stocks previously downloaded using yfinance
tickers = ["AAPL", "MSFT", "AMZN", "GOOGL", "NVDA", "META"]

# Raw data before reshaping
raw = yf.download(
    tickers=tickers,
    start=start_date,
    end=end_date,
    auto_adjust=False, # includes dividends and splits
    progress=False, # suppresses download progress bar
    group_by="ticker"
)

# Save raw data to CSV
raw.to_csv("data/stock_data.csv")
print(f"Raw data saved to stock_data.csv")
```
